```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
file_url = 'https://media.githubusercontent.com/media/musthave-ML10/data_source/main/car.csv'
```

```python
data = pd.read_csv(file_url)
data
```

```python
data['engine'].str.split(expand=True)
```

```python
data[['engine','engine_unit']] = data['engine'].str.split(expand=True)
```

```python
data['engine']= data['engine'].astype('float32')
```

```python
data['engine_unit'].value_counts()
```

```python
data.drop('engine_unit',axis=1,inplace=True)
```

```python
data['max_power']
```

```python
data[['max_power','max_power_unit']] = data['max_power'].str.split(expand=True)
```

```python
data['max_power'] = data['max_power'].astype('float32')
```

```python
data[data['max_power']=='bhp']
```

```python
def isFloat(val) :
    try :
        num = float(val)
        return num
    except ValueError:
        return np.NaN
```

```python
data['max_power'] = data['max_power'].apply(isFloat)
```

```python
data['max_power_unit'].unique()
```

```python
data.drop('max_power_unit',axis=1,inplace=True)
```

```python
data[['mileage','mileage_unit']]  = data['mileage'].str.split(expand=True)
```

```python
data['mileage'] = data['mileage'].astype('float32')
```

```python
data['mileage_unit'].value_counts()
```

```python
data['fuel'].unique()
# Petrol : 리터당 $80.43
# Diesel : 리터당 $73.56
# LPG : 킬로그램당 $40.85
# CNG : 킬로그램당 $44.23
```

```python
def mile(x) :
    if x['fuel'] == 'Pertrol' :
        return x['mileage'] / 80.43
    elif x['fuel'] == 'Diesel':
        return x['mileage'] / 73.56
    elif x['fuel'] == 'LPG' :
        return x['mileage'] / 40.85
    else :
        return x['mileage'] / 44.23
```

```python
data['mileage'] = data.apply(mile,axis=1)
```

```python
data.drop('mileage_unit',axis=1,inplace=True)
```

```python
data.head()
```

```python
data['torque'] = data['torque'].str.upper()
```

```python
def torque_unit(x) :
    if 'NM' in str(x) :
        return 'NM'
    elif 'KGM' in str(x) :
        return 'KGM'
```

```python
data['torque_unit'] = data['torque'].apply(torque_unit)
```

```python
data
```

```python
data['torque_unit'].value_counts()
```

```python
data[data['torque_unit'].isnull()]['torque'].unique()
```

```python
data['torque_unit'] = data['torque_unit'].fillna('NM')
```

```python
data['torque']
```

```python
def split_num(x) :
    x = str(x)
    for i,j in enumerate(x) :
        if j not in '0123456789.' :
            cut = i
            break
    return x[:cut]
```

```python
data['torque'] = data['torque'].apply(split_num)
```

```python
data
```

```python
data['torque']=data['torque'].astype('float32')
```

```python
data['torque']=data['torque'].replace('',np.NaN)
```

```python
data['torque']=data['torque'].astype('float32')
```

```python
data['torque'].head()
```

```python
def torque_tran(x) :
    if x['torque_unit'] == 'KGM' :
        return x['torque'] *9.8066
    else :
        return x['torque']
```

```python
data['torque'] = data.apply(torque_tran,axis=1)
```

```python
data.drop('torque_unit',axis=1,inplace=True)
```

```python
data.head()
```

```python
data['name'] = data['name'].str.split(expand=True)[0]
```

```python
data['name'].unique()
```

```python
data['name'] = data['name'].replace('Land','Land Rover')
```

```python
data.isnull().mean()
# 결측치가 약 2~3% 정도의 수준으로 보이고 있다.
# 결측치의 비율 2% 수준으로 높지 않기 때문에 과감하게 행을 모두 제거한다. 
```

```python
data.dropna(inplace=True)
len(data)
```

```python
data = pd.get_dummies(data,columns=['name','fuel','seller_type','transmission','owner'],drop_first=True)
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(data.drop('selling_price',axis=1),data['selling_price'],test_size=0.2,random_state=100)
```

```python
from sklearn.ensemble import RandomForestRegressor

model  = RandomForestRegressor(random_state=100)
model.fit(X_train,y_train)
train_pred = model.predict(X_train)
test_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error

print(mean_squared_error(y_train,train_pred)**0.5)
print(mean_squared_error(y_test,test_pred)**0.5)
```

```python
from sklearn.model_selection import KFold

data
```

```python
data.reset_index(drop=True,inplace=True)
```

```python
data
```

```python
kf = KFold(n_splits=5)
X = data.drop('selling_price',axis=1)
y = data['selling_price']
```

```python
for i,j in kf.split(X) :
    print(i,j)
```

```python
total_train_mse = []
total_test_mse = []
for train_idx , test_idx in kf.split(X) :
    X_train, X_test = X.loc[train_idx],X.loc[test_idx]
    y_train, y_test = y.loc[train_idx],y.loc[test_idx]


    model = RandomForestRegressor(random_state=100)
    model.fit(X_train,y_train)
    train_pred = model.predict(X_train)
    test_pred = model.predict(X_test)

    train_mse = mean_squared_error(y_train,train_pred) ** 0.5
    test_mse = mean_squared_error(y_test,test_pred) ** 0.5

    total_train_mse.append(train_mse)
    total_test_mse.append(test_mse)
```

```python
np.mean(total_train_mse)
```

```python
np.mean(total_test_mse)
```

```python
total_train_mse = []
total_test_mse = []

kf = KFold(n_splits=8)

for i,j in kf.split(X) :
    X_train,X_test = X.loc[i,:],X.loc[j,:]
    y_train,y_test = y.loc[i],y.loc[j]

    model = RandomForestRegressor(n_estimators=300,
                                  max_depth=50,
                                  min_samples_split=5,
                                  min_samples_leaf=1,
                                  n_jobs=-1,
                                  random_state=100)
    model.fit(X_train,y_train)
    train_pred = model.predict(X_train)
    test_pred = model.predict(X_test)

    train_mse = mean_squared_error(y_train,train_pred) ** 0.5
    test_mse = mean_squared_error(y_test,test_pred) ** 0.5

    total_train_mse.append(train_mse)
    total_test_mse.append(test_mse)
    
```
