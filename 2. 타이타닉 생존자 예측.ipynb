```python
file_url = 'https://media.githubusercontent.com/media/musthave-ML10/data_source/main/titanic.csv'
data = pd.read_csv(file_url)
data
```

```python
data = data.drop(['Name','Ticket'],axis=1)
```

```python
data.head()
```

```python
pd.get_dummies(data,columns=['Sex','Embarked']) # One Hot Encoding ì§„í–‰
```

```python
data = pd.get_dummies(data,columns=['Sex','Embarked'],drop_first=True) # One Hot Encoding ì§„í–‰
```

```python
data
```

```python
from sklearn.model_selection import train_test_split

X = data.drop('Survived',axis=1)
y = data['Survived']

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=100)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(X_train,y_train)
```

```python
pred = lr.predict(X_test)
pred
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test,pred) 
# ì´ì§„ë¶„ë¥˜ì´ê¸° ë•Œë¬¸ì— RMSEëŠ” í‰ê°€ì— ì í•©í•˜ì§€ ëª»í•¨
# ì´ì§„ë¶„ë¥˜ í‰ê°€ ì§€í‘œë¡œëŠ” ì •í™•ë„, ì˜¤ì°¨í–‰ë ¬, ì •ë°€ë„, ìž¬í˜„ìœ¨, F1 score, ë¯¼ê°ë„, íŠ¹ì´ë„, AUC ë“±ì´ ìžˆë‹¤.
```
