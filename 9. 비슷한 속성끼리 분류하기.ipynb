```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_url = 'https://raw.githubusercontent.com/musthave-ML10/data_source/main/example_cluster.csv'
data = pd.read_csv(file_url)
data # 데이터셋 불러오고 출력
```

```python
km.predict(data)
```

```python
data['label'] = km.predict(data)
```

```python
sns.scatterplot(x='var_1',y='var_2',data=data,hue='label',palette='rainbow')
```

```python
km.inertia_ # 이너셔 : 각 그룹에서의 중심과 각 그룹에 해당하는 데이터 간의 거리에 대한 합
```

```python
distance = []
for i in range(2,10) :
    model = KMeans(n_clusters=i)
    model.fit(data)
    distance.append(model.inertia_)
```

```python
file_url = 'https://raw.githubusercontent.com/musthave-ML10/data_source/main/customer.csv'
data = pd.read_csv(file_url)
data # 데이터셋 불러오고 출력
```

```python
data['cc_num'].nunique()
```

```python
data['category'].nunique()
```

```python
customer = pd.get_dummies(data,'category')
```

```python
customer.head()
```

```python
cat_list = customer.columns[2:]
cat_list
```

```python
for i in cat_list :
    customer[i] = customer[i] * customer['amt']
```

```python
customer
```

```python
customer_agg = customer.groupby('cc_num').sum() # cc_num별 총 금액
customer_agg.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_df = pd.DataFrame(scaler.fit_transform(customer_agg),
                         columns=customer_agg.columns,
                         index=customer_agg.index)
scaled_df
```

```python
distance = []

for i in range(2,10) :
    model = KMeans(n_clusters=i)
    model.fit(scaled_df)
    distance.append(model.inertia_)
```

```python
plt.plot(range(2,10),distance)
```

```python
from sklearn.metrics import silhouette_score

silhouette = []
for i in range(2,10) :
    model = KMeans(n_clusters=i)
    model.fit(scaled_df)
    labels = model.predict(scaled_df)
    silhouette.append(silhouette_score(scaled_df,labels))
```

```python
sns.lineplot(x=range(2,10),y=silhouette) # 4일 때 가장 좋은 분류 성능을 나타낸다.
```

```python
k_model = KMeans(n_clusters=4)
k_model.fit(scaled_df)
labels = k_model.predict(scaled_df)
```

```python
scaled_df['label'] = labels
scaled_df_mean = scaled_df.groupby('label').mean()
scaled_df_count = scaled_df.groupby('label').count()['category_travel']
```

```python
scaled_df_count = scaled_df_count.rename('count')

scaled_all = scaled_df_mean.join(scaled_df_count)

scaled_all
```
